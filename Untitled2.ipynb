```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("open/train_features.csv")
train_label = pd.read_csv("open/train_labels.csv")
test_data = pd.read_csv("open/test_features.csv")

test_data
```

```python
train_merge = pd.merge(train_data,train_label,on="id",how="outer")
train_merge_sorted = train_merge.sort_values(by=['label','id'],ascending=True)
```

```python
print(train_merge_sorted.value_counts('label')/600)
print("총 합",train_merge_sorted.value_counts('label').sum()/600)

# Non-Exercise == 26, 운동 안하는 기록이 3125개 데이터중, 1518개나 됨;;
```

```python
#188번째 사람 운동 패턴

test_118 = train_merge_sorted[train_merge_sorted['id'] == 118]
test_118.drop(['time','id','label'],axis=1,inplace=True)
print(test_118.plot())
```

```python
# 12번 운동을 한 사람들 id

train_merge_sorted[train_merge_sorted['label'] == 12]['id'].value_counts().index.to_list()
```

```python
#라벨별 기록된 데이터(초당) 개수 -> /600 해주면 하나의 데이터 개수 나옴
%matplotlib inline
list_label = train_merge_sorted['label'].value_counts()
list_label_index = list_label.index
print(list_label)
print("총 개수 :",list_label.sum())
```

```python
colnames = train_merge_sorted.columns
colnames = colnames.drop(['time','id','label','label_desc'])
colnames
```

```python
from tqdm import tqdm
from sklearn.preprocessing import minmax_scale

result = pd.DataFrame(columns=colnames) # 빈 데이터 프레임

for i in tqdm(list_label.index):
    dummy = train_merge_sorted[train_merge_sorted['label'] == i]
    dummy.drop(['time','id','label','label_desc'],axis=1,inplace=True)
    
    
    # 데이터 정규화
    dummy = minmax_scale(dummy)
    dummy = pd.DataFrame(dummy,columns=colnames)
        
    for c in tqdm(dummy.columns.to_list()):
        for i in dummy[c]:
            if i < dummy[c].quantile(0.05):
                dummy[dummy[c]==i] = dummy[c].quantile(0.05)
            elif i > dummy[c].quantile(0.95):
                dummy[dummy[c]==i] = dummy[c].quantile(0.95)
                
    result = pd.concat([result,dummy], ignore_index=True)
    #dummy.plot()
    
    


# 5만개 돌리는데 약 2~3분, 1만개당 0.5분, 총 180만개 데이터 -> 90분 예상;
# 데이터를 몇개만 뽑아서 처리 하기 vs 이렇게 접근하는 방법이 아니다.
# 전자를 선택
# 뽑을 데이터는 agg 를 이용해, 평균보다 상 하위 90%, 10% 차이 나는 데이터만 뽑아보기
```

```python
features = ['id', 'acc_x', 'acc_y', 'acc_z', 'gy_x', 'gy_y', 'gy_z']
X_train = train_data[features].groupby('id').agg(['max', 'min', 'mean'])
X_test = test_data[features].groupby('id').agg(['max', 'min', 'mean'])

y_train = train_label['label']

X_test
```

```python
X_train.shape, X_test.shape

# 실패! 처음부터 접근이 잘못됨, 어차피 상 하위 90, 10% 초과 미만 애들 찾을라 해도, 
# 전체 데이터를 돌려바야돼서 똑같음;
# 전체 데이터를 돌려서 전처리 하는 방법은 없는듯. -> 너무 비효율적
```

```python
# ransac에 x축으로 넣어줄 min, max값 구하기 위한 전처리 작업(모든 데이터 중, 숫자만
# 뽑아와 최소, 최대값 구함)


save = [str(X_train).split(' ')]
save_data = []
for i in save[0]:
    try:
        save_data.append(float(i))
    except TypeError:
        pass
    except ValueError:
        pass
    
print(save_data)
print("\n\n\n min : ", min(save_data), "max : ", max(save_data))
```

```python
#오리지날 데이터 그래프 모양
plt.plot(X_train, y_train)
```

```python
# ransac

import numpy as np
from matplotlib import pyplot as plt

from sklearn import linear_model


# Fit line using all data
lr = linear_model.LinearRegression()
lr.fit(X_train, y_train)

# Robustly fit linear model with RANSAC algorithm
ransac = linear_model.RANSACRegressor()
ransac.fit(X_train, y_train)
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)

# Predict data of estimated models
line_X = np.arange(min(save_data),max(save_data))[:, np.newaxis]
line_y = lr.predict(X_test)
line_y_ransac = ransac.predict(X_test)

line_y_test = ransac.predict(X_train)

# Compare estimated coefficients
# print("Estimated coefficients (true, linear regression, RANSAC):")
# print(coef, lr.coef_, ransac.estimator_.coef_)

# lw = 2
# plt.scatter(X_train[inlier_mask], y_train[inlier_mask], color='yellowgreen', marker='.',
#             label='Inliers')
# plt.scatter(X_train[outlier_mask], y_train[outlier_mask], color='gold', marker='.',
#             label='Outliers')

plt.plot(X_test, line_y, color='navy', label='Linear regressor')
plt.plot(X_test, line_y_ransac, color='cornflowerblue',
         label='RANSAC regressor')
plt.legend(loc='lower right')
plt.xlabel("Input")
plt.ylabel("Response")
plt.show()
```
